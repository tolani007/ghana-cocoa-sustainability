```python

fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# Rainfall vs Yield
ax[0].scatter(merged["Rainfall_mm"], merged["Yield_kg_per_ha"])
ax[0].set_title("üåßÔ∏è Rainfall vs Cocoa Yield")
ax[0].set_xlabel("Rainfall (mm)")
ax[0].set_ylabel("Yield (kg/ha)")

# Temperature vs Yield
ax[1].scatter(merged["AvgTemp_C"], merged["Yield_kg_per_ha"])
ax[1].set_title("üå°Ô∏è Temperature vs Cocoa Yield")
ax[1].set_xlabel("Avg Temperature (¬∞C)")
ax[1].set_ylabel("Yield (kg/ha)")

plt.tight_layout()
plt.show()

```

### üå¶Ô∏è Rainfall and üå°Ô∏è Temperature vs Cocoa Yield

```python
import seaborn as sns

plt.figure(figsize=(6, 5))
sns.heatmap(merged.corr(numeric_only=True), annot=True, cmap="YlGnBu")
plt.title("üìç Correlation Matrix (1961‚Äì2023)")
plt.show()
```
